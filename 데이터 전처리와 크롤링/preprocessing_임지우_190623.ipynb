```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline 
```

```python
bugs = pd.read_csv('Bugs.csv')
bugs.head()
```

1. singer에서 "connect 아티스트" 문구제거

```python
for i in bugs[bugs['singer'].str.contains('CONNECT 아티스트')].index:
    bugs.loc[i,'singer'] = bugs.loc[i,'singer'][:-13]
bugs.head()
```

2. 필요한 연도만 뽑기

```python
erase_list = ["1940's 이전 베스트 가요 콜렉션","1950's 베스트 가요 콜렉션","1960's 베스트 가요 콜렉션",
             "1960~1963년 베스트 가요 콜렉션","1970's 베스트 가요 콜렉션","1980's 베스트 가요 콜렉션",
             "1990's 베스트 가요 콜렉션","2000's 베스트 가요 콜렉션"]
```

```python
bugs = bugs[~bugs['year'].isin(erase_list)]
bugs.tail()
```

```python
bugs['year'] = bugs['year'].apply(lambda x: x[:4])
bugs.head()
```

3. runtime 초 단위로 변환

```python
revised_time_list = []

for item in bugs['runtime']:
    splitted = item.split(':')
    minute = int(splitted[0][1])*60
    seconds = int(splitted[1])
    revised_time = minute+seconds
    revised_time_list.append(revised_time)
    
bugs['runtime'] = revised_time_list
bugs.head()
```
