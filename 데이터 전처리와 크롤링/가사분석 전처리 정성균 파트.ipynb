## 각각에 대한 처리
1. 대한민국
2. 솔로
3. 그룹
4. 미국

```python
groupma = ["올스타 밴드","에밀레","야생마","홍삼트리오","템페스트","휘버스 (열기들)","샌드 페블즈(Sand Pebbless)"]
groupfe = ["쿨 시스터즈","이 시스터즈","정 시스터즈","이씨스터즈","화니 시스터즈(Fanny Sisters)"]
grouphon = ["이건주와 어린이들","하눌타리","E.G.B","퓨처라이거"]
duoma = ["하사와 병장"]
duofe = ["작품 하나"]
soloma = ["하성관","이예성","유춘산","도성아","이신","정연택","신세영","한정무","허민"]
solofe = ["나현희","정난이","허림","조정희","정은숙","이정민","심연옥"]
```

```python
for i in range(len(bugs)) :
    a = bugs[i:i+1]
    if a.singer.item() in groupma :
        bugs.iloc[i]["type"] = "그룹 (남성)"
    if a.singer.item() in groupfe :
        bugs.iloc[i]["type"] = "그룹 (여성)"
    if a.singer.item() in grouphon :
        bugs.iloc[i]["type"] = "그룹 (혼성)"
    if a.singer.item() in duoma :
        bugs.iloc[i]["type"] = "듀오 (남성)"
    if a.singer.item() in duofe :
        bugs.iloc[i]["type"] = "듀오 (여성)"
    if a.singer.item() in soloma :
        bugs.iloc[i]["type"] = "솔로 (남성)"
    if a.singer.item() in solofe :
        bugs.iloc[i]["type"] = "솔로 (여성)"
```
